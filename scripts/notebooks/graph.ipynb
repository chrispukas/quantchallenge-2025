```python
import os
import pandas as pd
from importlib.resources import files

import qch2025.pkg.plotting as pltt
import matplotlib.pyplot as plt

datasets = files("qch2025.dataset")

train_csv: str = datasets.joinpath("train.csv")
test_csv: str = datasets.joinpath("test.csv")

df: pd.DataFrame = pd.read_csv(train_csv)
x = df["A"]
y = df["Y1"]
pltt.plot_scatter(x, y, x_label="A", y_label="Y1", title="A vs Y1")

x = df["A"]
y = df["Y2"]
pltt.plot_scatter(x, y, x_label="A", y_label="Y2", title="A vs Y2")

m = df["N"]
t = df["time"]

plt.plot(t, m)
```
